## 데이터프레임 창조

```python
pf = pd.DataFrame(data = [[25, 35, 8, 18],
                          [18, 27, 10, 20],
                          [17, 17, 10, 19],
                          [12, 22, 9, 20],
                          [22, 35, 8, 16]],
                  index=['윤일형', '강수화', '홍소희', '유한빈', '신수빈'],
                  columns=['중간', '기말', '과제', '출석'])
pf
```

```python
pf['중간']
```

```python
pf.중간
```

```python
pf['출석']
```

```python
pf[['중간']]
```

```python
pf[['기말', '출석', '중간']]
```

```python
pf[['기말']]
```

```python
pf['중간'] > 20
```

```python
pf[pf['중간'] > 20]
```

```python
pf[pf['출석'] >= 19]
```

```python
pf[(pf['출석'] >= 19) & (pf['기말'] >= 20)]
```
